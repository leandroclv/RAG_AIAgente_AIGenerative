####While Loops

It's a repeating cycle that executes a block of code continuously while a specific condition remains true.

The loop runs while the condition is true,and when it becmes false, the loop terminates, and the program continues to the next line of code outside the loop

```python
#while condition:
#    code to be executed
```

```python
people_in_car = 0

while people_in_car < 5:
    print(f'People in the car: {people_in_car}')
    people_in_car += 1
    
print('No more people allowed.')
```

```python
number = 0

while number <= 20:
    if number % 2 == 0:
        print(number, end =' ')
    number += 1
```

```python
#Countdown times
countdown = 10

while countdown >= 0:
    if countdown == 0:
        print('Liftoff')
    else:
        print(f'T-minus {countdown} seconds')
    countdown -= 1
```

###Lists

- Python lists allows to store multiple items in a single variable
- They're ordered, mutable, and can include various data types
- To create a list
    - put items within square brackets
    - separate them with commas

```python
#my_list = [item1, item2, item3, ...]
numbers = [1,2,3,4,5]
mixed_list = [42, 'hello', 3.14, True]
```

Get individual list items using index, starting from 0 for the first item.

```python
print(numbers[0]) #Output: 1
print(mixed_list[2]) #Output: 3.14
```

Slicing -> to get a portion of a list, specify a start and end index

```python
print(numbers[1:4])
```

Modifying items -> you can change list contents by assigning a new value to the corresponding index

```python
numbers[0] = 100
print(numbers)
```

Adding items -> you can append items to a list using the append() method or the + operator

```python
numbers.append(6)
print(numbers)

numbers = numbers + [7,8,9]
print(numbers)
```

```python
del numbers[0]
print(numbers)

numbers.remove(7)
print(numbers)

last_number = numbers.pop()
print(numbers)
print(last_number)
```
